Actualmente la educación en el país Mexicano ha sido uno de los mayores problemas psicosociales en la actualidad, ya que un bajo nivel educativo refleja un bajo nivel en el desarrollo del país y asimismo, un bajo nivel de especialización y de buenas oportunidades laborales, llevando a la pobreza. Ejemplo de lo explicado anteriormente es el bajo desempeño que los estudiantes mexicanos demostraron durante la prueba del Programa para la Evaluación Internacional de Alumnos de la OCDE (PISA por sus siglas en ingles). La evaluación PISA es un examen aplicado a países que conforman la OCDE que evalua preguntas en torno a la lectura, matemáticas, ciencias y competencias globales como áreas menores de evaluación. En la pasada Evaluación PISA del 2018, México llegó a estar en el tercer puesto de los 4 países latinoamericanos que se les aplico la prueba e incluso llegando a estare en el penúltimo lugar a nivel global en el área de lectura y a figurar en puestos bajos en las demas áreas. 

Las estadísticas anteriores resultan ser aterradoras para un país en vías de desarrollo, ya que la falta de un buen sistema educativo frena la innovación y desarollo de un país. Una posible solución al problema del pésimo desempeño en esas zonas, puede ser la practica constante. Mediante la constante practica se puede agudizar el conocimiento en los temas, facilitando asi que los jóvenes pueden ahora enfrentarse a problemas similares ya que obtienen un grado de maestria en la materia. Asimismo, para evitar que las preguntas se reciclen y sean las mismas, una propuesta es crear un banco de preguntas dinámico, al cual se le puedan insertar preguntas similares a las de los exámenes y que permitan que, en caso de que un alumno ya haya visto y resulto todas las preguntas que el examen contenia en modo de practica, pueda seguir aprendiendo de los diferentes temas.

```python
def registrar_pregunta(nombre): #Funcion que permite resgistrar preguntas en el banco de preguntas al pedir un ID y todos los datos que conforman una pregunta y termina siendo añadida a la matriz de preguntas


  i = 0
  j = 0

  leer_archivo(nombre)

  for pregunta in preguntas:
    i += 1

  j = i +1
  
  valor_registro = input("Desea registrar una pregunta? (si / no) ")

  if valor_registro == "si" or valor_registro == "Si":

    id_pregunta = input(f"Ingrese ID de la pregunta (Ultimo ID registrado {i} (Id recomendado {j}))")
    enfoque = str(input("Favor de insertar el enfoque de la pregunta (lectura, matemáticas y ciencias): "))
    pregunta = str(input("Favor de insertar la pregunta: "))
    respuestas = ""

    for num in range(4):

      respuesta = input("Ingresa la respuesta " + str(num + 1) + ": ")

      if num == 3:
        respuestas = respuestas + respuesta 

      else:
        respuestas = respuestas + respuesta + "," 

    respuestas = respuestas.split(",")

    respuesta1, respuesta2, respuesta3, respuesta4 = respuestas 

    correcta = input("Ingresa la respuesta correcta:")

    pregunta = [id_pregunta, enfoque, pregunta, respuesta1, respuesta2, respuesta3, respuesta4, correcta]
    
    preguntas.append(pregunta)

    guardar_archivo(nombre_archivo, preguntas)

  else:
    print("Volviendo a menu")

registrar_pregunta(nombre_archivo)
```

```python
def actualizar_pregunta(nombre): #Permite modificar una pregunta al pedir el ID, el valor que se desea modificar y la modificacion. Termina ingresando la nueva pregunta en la lista

  global nombre_archivo
  global preguntas

  valor = 1
  i=0
  preguntas_actualizadas = []


  leer_archivo(nombre_archivo)
  
  print("Area de actualizacion de preguntas \nA continuación se despliega la lista de preguntas guardadas\n")

  for pregunta in preguntas:    
    id_pregunta, enfoque, pregunta, respuesta1, respuesta2, respuesta3, respuesta4, correcta = pregunta 

    print(f"ID de la pregunta: {id_pregunta} \nEnfoque de la pregunta: {enfoque} \nPregunta: {pregunta}")
    print(f"a){respuesta1:<15}b){respuesta2:<15}c){respuesta3:<15}d){respuesta4:<15} \n")

    valor += 1 

  while True: 
    
    id_cambio = input("Ingresa el id de la pregunta que quieres cambiar (Si desea salir ingrese x): ")

    if id_cambio == "x":
      break
   
    elif int(id_cambio) < valor:

      for pregunta in preguntas:

        id_pregunta, enfoque, pregunta, respuesta1, respuesta2, respuesta3, respuesta4, correcta = pregunta 
        preg = [id_pregunta, enfoque, pregunta, respuesta1, respuesta2, respuesta3, respuesta4, correcta]  
        
        if id_pregunta == id_cambio:

          print(f"ID de la pregunta: {id_pregunta} \nEnfoque de la pregunta: {enfoque} \nPregunta: {pregunta}")
          print(f"a){respuesta1:<15}b){respuesta2:<15}c){respuesta3:<15}d){respuesta4:<15} \n")

          while True:

            parametro = input("Ingrese que dato desea cambiar(enfoque, pregunta, respuesta1, respuesta2, respuesta3, respuesta4, correcta) o si desea salir teclee x ")

            if parametro == "x":
              break

            elif parametro == "enfoque" or parametro == "pregunta" or parametro == "respuesta1" or parametro == "respuesta2" or parametro == "respuesta3" or parametro == "respuesta4" or parametro == "correcta":
                
              cambio = input("Ingrese el cambio")

              if parametro == "enfoque":
                 enfoque = cambio

              elif parametro == "pregunta":
                pregunta = cambio

              elif parametro == "respuesta1":
                respuesta1 = cambio

              elif parametro == "respuesta2":
                respuesta2 = cambio
                
              elif parametro == "respuesta3":
                respuesta3 = cambio

              elif parametro == "respuesta4":
                respuesta4 = cambio

              elif parametro == "correcta":
                correcta = cambio

              while True:

                confirmacion = input("Esta seguro que desea guardar los cambios?(si / no)")

                if confirmacion == "si":
                  pregunta = [id_pregunta, enfoque, pregunta, respuesta1, respuesta2, respuesta3, respuesta4, correcta]
                  preguntas_actualizadas.append(pregunta)
                  break
                
                elif confirmacion == "no":
                  preg_renovado = [id_pregunta, enfoque, pregunta, respuesta1, respuesta2, respuesta3, respuesta4, correcta]
                  preguntas_actualizadas.append(preg_renovado)

                else:
                  print("Seleccione una opcion valida")

              print("La pregunta actualizada es: ")
              id_pregunta, enfoque, pregunta, respuesta1, respuesta2, respuesta3, respuesta4, correcta = pregunta 

              print(f"ID de la pregunta: {id_pregunta} \nEnfoque de la pregunta: {enfoque} \nPregunta: {pregunta}")
              print(f"a){respuesta1:<15}b){respuesta2:<15}c){respuesta3:<15}d){respuesta4:<15} \n")

            else:
              print("Ingrese un parametro valido")  

        else: 
          preg_renovado = [id_pregunta, enfoque, pregunta, respuesta1, respuesta2, respuesta3, respuesta4, correcta]
          preguntas_actualizadas.append(preg_renovado) 

        preguntas = preguntas_actualizadas         
              
    else: 
      print("Ingrese ID valido")

    guardar_archivo(nombre_archivo, preguntas)

actualizar_pregunta(nombre_archivo) 
```
